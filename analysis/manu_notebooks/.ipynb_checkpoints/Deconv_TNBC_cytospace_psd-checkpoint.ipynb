```python
ad_xen_sp = sc.read_h5ad(data_path+'TNBC_Xenium_spot_anno.h5ad')
ad_xen_sp
```

```python
ad_xen_sp = ad_xen_sp[ad_xen_sp.obs['spot'].isin(ad_vis.obs.index)]
ad_xen_sp
```

```python
ad_xen_sp.obs['cell_type'].value_counts()/ad_xen_sp.obs['cell_type'].value_counts().sum()
```

```python
deconv_df = ad_xen_sp.obs.pivot_table(index='spot', columns='cell_type', aggfunc='size').reindex(ad_vis.obs.index).T
deconv_df = deconv_df/deconv_df.sum(axis=0)
deconv_df
```

```python
vis_decon = celltype_portion
vis_decon = vis_decon[['epithelial', 'Immune', 'stroma']]
vis_decon.columns = ['Epithelial', 'Immune', 'Stroma']
```

```python
psr = pd.DataFrame()
psr.loc['psr_cytospace', 'Epithelial'] = stats.pearsonr(deconv_df.T['Epithelial'], vis_decon['Epithelial'])[0]
psr.loc['psr_cytospace', 'Immune'] = stats.pearsonr(deconv_df.T['Immune'], vis_decon['Immune'])[0]
psr.loc['psr_cytospace', 'Stroma'] = stats.pearsonr(deconv_df.T['Stroma'], vis_decon['Stroma'])[0]
psr
```

```python
ad_vis.obs = pd.concat([ad_vis.obs, celltype_portion],axis=1)
ad_vis.obs
```
