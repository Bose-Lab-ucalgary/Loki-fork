```python
ad_xen_sp.obs['cell_type'].value_counts()/ad_xen_sp.obs['cell_type'].value_counts().sum()
```

```python
deconv_df = ad_xen_sp.obs.pivot_table(index='spot', columns='cell_type', aggfunc='size', fill_value=0).reindex(ad_expr.obs.index).T
deconv_df = deconv_df/deconv_df.sum(axis=0)
deconv_df
```

```python
raw_port = deconv_df.T
raw_port
```

```python
vis_decon = celltype_portion.copy()
vis_decon
```

```python
vis_decon['Immune'] = vis_decon['macrophage'] + vis_decon['Bcell'] + vis_decon['Tcell']
vis_decon.drop(columns=['macrophage', 'Bcell', 'Tcell','endothelial'], inplace=True)
vis_decon = vis_decon[['epithelial', 'Immune', 'stroma']]
```
